```python
import dataset_module as DS
import distance_module as Distance
import statistics_module as Stati
```

```python
# Function for user interface
def main():
    # Prompt the user to select an option
    while True:
        print("\nWelcome to the transaction analysis system!")
        print("Please select an option:")
        print("1. Compute the distance between two transactions of a user.")
        print("2. Compute the distance between two users' transactions.")
        print("3. Compute the average transactions of a user.")
        print("4. Compute the average transactions of all users.")
        print("5. Compute the mode of transactions of a user.")
        print("6. Compute the mode of transactions of all users.")
        print("7. Compute the median of transactions of a user.")
        print("8. Compute the median of transactions of all users.")
        print("9. Compute the interquartile range of a user's transactions.")
        print("10. Compute the interquartile range of all users' transactions.")
        print("11. Compute the location centroid of a user.")
        print("12. Compute the standard deviation of a user's transactions.")
        print("13. Check whether a transaction is fraudulent.")
        print("14. Return abnormal transactions for a user.")
        print("15. Compute the Z score of a user's transactions.")
        print("16. Compute the Z score of all users' transactions.")
        print("17. Compute the frequency of transactions at a given location.")
        print("18. Check the outlier location of a user.")
        print("19. Compute the nth percentile of transactions of a user.")
        print("20. Compute the nth percentile of transactions of all users.")
        print("0. Exit")
        option = input("Option: ")

        # Call the appropriate function based on the user's choice
        if option == "1":
            uid = input("Enter the user ID: ")
            tran1 = input("Enter the first transaction ID: ")
            tran2 = input("Enter the second transaction ID: ")
            distance = Distance.SameUserDistance(uid,tran1, tran2)
            print("The distance between transactions " + str(tran1) + " and " + str(tran2) + " is " + str(round(distance, 2)) + ".")
        elif option == "2":
            user1 = input("Enter the first user ID: ")
            user2 = input("Enter the second user ID: ")
            tran1 = input("Enter the first users' transaction ID: ")
            tran2 = input("Enter the second users' transaction ID: ") 
            distance = Distance.TwoUserDistance(user1, user2, tran1, tran2)
            print("The distance between transactions " + str(tran1) + " of user " + str(user1) +" and " + str(tran2) +" of user " + str(user2) + " is " + str(round(distance, 2)) + ".")

        elif option == "3":
            user = input("Enter the user ID: ")
            avg = Stati.Mean(user)
            print("The average transactions of user " + str(user) + " is " + str(round(avg, 2)) + ".")
        elif option == "4":
            avg = Stati.Mean()
            print("The average transactions of all users is " + str(round(avg, 2)) + ".")


        elif option == "5":
            user = input("Enter the user ID: ")
            mode = Stati.Mode(user)
            print("The mode of transactions of user " + str(user) + " is " + str(mode) + ".")
        elif option == "6":
            mode = Stati.Mode()
            print("The mode of transactions of all users is " + str(mode) + ".")

        elif option == "7":
            user = input("Enter the user ID: ")
            median = Stati.Median(user)
            print("The median of transactions of user " + str(user) + " is " + str(round(median, 2)) + ".")    
        elif option == "8":
            median = Stati.Median()
            print("The median of transactions of all users is " + str(round(median, 2)) + ".")


        elif option == "9":
            user = input("Enter the user ID: ")
            interquartile_range = Stati.IQR(user)
            print("The interquartile range of user " + str(user) + "'s transactions is " + str(round(interquartile_range, 2)) + ".")
        elif option == "10":
            interquartile_range = Stati.IQR()
            print("The interquartile range of all users' transactions is " + str(round(interquartile_range, 2)) + ".")

        elif option == "11":
            user = input("Enter the user ID: ")
            centroid = Stati.LocCentroid(user)
            print("The location centroid of user " + str(user) + " is " + str(centroid) + ".")

        elif option == "12":
            user = input("Enter the user ID: ")
            stdev = Stati.SD(user)
            print("The standard deviation of user " + str(user) + "'s transactions is " + str(round(stdev, 2)) + ".")

        elif option == "13":
            transaction = input("Enter the transaction ID: ")
            is_fraudulent = Stati.fraudOrNot(transaction)
            print("The transaction " + str(transaction) + " is " + str(is_fraudulent) + ".")


        elif option == "14":
            user = input("Enter the user ID: ")
            print("The abnormal transactions for user " + str(user) + " are " + ":")
            abnormal_transaction = Stati.abnormal(user)

        elif option == "15":
            user = input("Enter the user ID: ")
            print("The Z score of user " + str(user) + "'s transactions are " + ":")
            z_score = Stati.ZScore(user)


        elif option == "16":
            print("The Z score of all users transactions are " + ":")
            z_score = Stati.ZScore()

        elif option == "17":
            x_coordinate  = float(input("Enter the locations' x coordinate : "))
            y_coordinate  = float(input("Enter the locations' y coordinate : "))
            frequency = Stati.Locfreq(x_coordinate,y_coordinate)
            print("The frequency of transactions at location " + str(x_coordinate) + " and " + str(y_coordinate) + " is " + str(frequency) + ".")


        elif option == "18":
            user = input("Enter the user ID: ")
            print("The outlier transactions for user " + str(user) + " are " + ":")
            outlier = Stati.Outlier(user)


        elif option == "19":
            user = input("Enter the user ID: ")
            percentile = float(input("Enter the percentile (between 0 and 100): "))
            nth_percentile = Stati.nthPercentile(percentile, user)
            print("The " + str(percentile) + "th percentile of transactions of user " + str(user) + " is " + str(round(nth_percentile, 2)) + ".")

        elif option == "20":
            percentile = float(input("Enter the percentile (between 0 and 100): "))
            nth_percentile = Stati.nthPercentile(percentile)
            print("The " + str(percentile) + "th percentile of transactions of all users is " + str(round(nth_percentile, 2)) + ".")
        elif option == "0":
            break
        else:
            print("Invalid option.")

```

```python
main()
```
